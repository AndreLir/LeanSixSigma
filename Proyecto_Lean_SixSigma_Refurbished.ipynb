##1.4 Kano Analysis basado en el VOC

| # | VOC (cita corta)                                              | Necesidad / Requerimiento              | CTQ asociado                     | Categor√≠a Kano            | Justificaci√≥n                                                                |
| - | ------------------------------------------------------------- | -------------------------------------- | -------------------------------- | ------------------------- | ---------------------------------------------------------------------------- |
| 1 | ‚ÄúTenemos sobrecostos por horas extra y contrataciones‚Äù (GE)   | Reducir costos de operaci√≥n            | Costo extra ‚â§ X%                 | Must-be                   | Los clientes internos esperan eficiencia de costos como condici√≥n m√≠nima.    |
| 2 | ‚ÄúEl problema es la variaci√≥n en tiempos de ciclo‚Äù (GO)        | Estabilidad en el flujo                | Desviaci√≥n est√°ndar ‚â§ Y d√≠as     | One-dimensional           | Mientras m√°s estable sea el ciclo, mayor satisfacci√≥n y confianza.           |
| 3 | ‚ÄúLos equipos entran irregulares a test‚Äù (GST)                 | Flujo continuo hacia test              | % de d√≠as con flujo estable > Z% | One-dimensional           | La mejora directa del flujo impacta en la entrega puntual y sin re-trabajos. |
| 4 | ‚ÄúNo siempre hay personal suficiente en limpieza/pintado‚Äù (LT) | Balancear recursos                     | % de d√≠as con ritmo esperado     | Must-be                   | La disponibilidad de personal es b√°sica, sin ella no puede ejecutarse nada.  |
| 5 | ‚ÄúNecesitamos cumplir 4800 ONT en 15 d√≠as sin extras‚Äù (GE)     | Cumplimiento de plazos sin sobretiempo | % de pedidos a tiempo ‚â• 95%      | Must-be / One-dimensional | Es una exigencia contractual y esperada. Si se cumple genera satisfacci√≥n.   |
| 6 | ‚ÄúNo tenemos visibilidad clara del ciclo‚Äù (PM s√≠ntesis)        | Medici√≥n y control del proceso         | Dashboard en tiempo real         | Attractive                | No siempre es pedido, pero un tablero visual genera valor y confianza extra. |



````python
# @title
import matplotlib.pyplot as plt
import numpy as np

# --------------------------
# Datos adaptados al caso ONT
# --------------------------
requirements = [
    "Reducir sobrecostos (horas extra/contrataciones)",  # GE
    "Estabilidad en tiempos de ciclo",                   # GO
    "Flujo continuo hacia test",                         # GST
    "Disponibilidad suficiente de personal",             # LT
    "Cumplir 4800 ONTs en 15 d√≠as sin extras",           # GE
    "Visibilidad clara del ciclo (dashboard en tiempo real)" # PM
]

categories = [
    "Must-be",         # 1: Costos
    "One-dimensional", # 2: Estabilidad ciclo
    "One-dimensional", # 3: Flujo a test
    "Must-be",         # 4: Personal
    "Must-be",         # 5: Plazos
    "Attractive"       # 6: Dashboard
]

# --------------------------
# Colores para las categor√≠as
# --------------------------
colors = {
    "Must-be": "red",
    "One-dimensional": "blue",
    "Attractive": "green",
    "Indifferent": "gray"
}

# --------------------------
# Centros m√°s separados para dispersi√≥n
# --------------------------
coords_center = {
    "Must-be": (2, 2), # Base coordinate for Must-be
    "One-dimensional": (6, 6),
    "Attractive": (9, 9),
    "Indifferent": (6, 2)
}

# --------------------------
# Generar dispersi√≥n aleatoria, adjusted for Must-be
# --------------------------
np.random.seed(42)  # Reproducible
scatter_coords = []
must_be_count = 0
for cat in categories:
    cx, cy = coords_center[cat]
    if cat == "Must-be":
        # Slightly adjust the base coordinates for each Must-be point
        x = cx + np.random.uniform(-0.6, 0.6)
        y = cy + np.random.uniform(-0.6, 0.6) + must_be_count * 0.5  # Increased y-offset
        must_be_count += 1
    else:
        x = cx + np.random.uniform(-0.6, 0.6)
        y = cy + np.random.uniform(-0.6, 0.6)
    scatter_coords.append((x, y))


# --------------------------
# Graficar
# --------------------------
plt.figure(figsize=(10, 7))
for (req, cat, (x, y)) in zip(requirements, categories, scatter_coords):
    plt.scatter(x, y, color=colors[cat], s=120,
                label=cat if cat not in plt.gca().get_legend_handles_labels()[1] else "")
    plt.text(x+0.25, y, req, fontsize=9)

# L√≠neas gu√≠a
plt.axhline(5, color="gray", linestyle="--", linewidth=0.8)
plt.axvline(5, color="gray", linestyle="--", linewidth=0.8)

# Configuraci√≥n
plt.title("Kano Analysis ‚Äì Proyecto Lean Six Sigma (ONT)", fontsize=14)
plt.xlabel("Nivel de desempe√±o del proceso/servicio", fontsize=11)
plt.ylabel("Nivel de satisfacci√≥n del cliente interno", fontsize=11)
plt.xlim(0, 11)
plt.ylim(0, 11)
plt.legend()
plt.show()
````

````python
# @title
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# ---------------------------
# 1. Definir la matriz RACI adaptada al proyecto ONT
# ---------------------------
data = {
    "Actividad / Entregable": [
        "Definir alcance del proyecto LSS",
        "Recolectar VOC de stakeholders",
        "Mapear SIPOC del proceso ONT",
        "Construcci√≥n dataset y m√©tricas CTQ (tiempo ciclo, retrasos)",
        "An√°lisis Kano y priorizaci√≥n de requerimientos",
        "Levantamiento de informaci√≥n en taller (limpieza, pintado, test)",
        "Seguimiento de tiempos de ciclo y productividad",
        "Informe de resultados a Direcci√≥n",
        "Plan de mejoras / estandarizaci√≥n de procesos"
    ],
    "Project Manager (LSS)": ["R","R","R","R","R","I","R","R","R"],
    "Gerente Operaciones": ["A","I","C","I","A","I","A","A","A"],
    "Gerente Soporte T√©cnico": ["C","C","C","I","C","A","C","C","C"],
    "Jefe de Taller": ["C","C","A","C","C","R","C","C","C"],
    "Gerente General": ["I","A","I","I","I","I","I","A","A"],
    "Equipo de Datos / QA": ["I","I","I","A","I","C","A","I","I"]
}

df = pd.DataFrame(data)

# ---------------------------
# 2. Codificar RACI a colores
# ---------------------------
color_map = {
    "R": "#4CAF50",   # Verde
    "A": "#2196F3",   # Azul
    "C": "#FFC107",   # Amarillo
    "I": "#E0E0E0",   # Gris
    "A/R": "#9C27B0"  # Morado (cuando tiene doble rol)
}

# Convertir cada celda a color
df_colors = df.iloc[:, 1:].map(lambda x: color_map.get(x, "white"))

# ---------------------------
# 3. Dibujar la matriz RACI
# ---------------------------
fig, ax = plt.subplots(figsize=(13, 6))

# Dibujar tabla con colores
table = plt.table(
    cellText=df.iloc[:, 1:].values,
    rowLabels=df["Actividad / Entregable"],
    colLabels=df.columns[1:],
    cellColours=df_colors.values,
    cellLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1.2, 1.5)

ax.axis('off')
plt.title("Matriz RACI ‚Äì Proyecto Lean Six Sigma (ONT)", fontsize=14, pad=20)

# Leyenda
legend_elements = [
    Patch(facecolor=color_map['R'], label='R: Responsable'),
    Patch(facecolor=color_map['A'], label='A: Aprobador'),
    Patch(facecolor=color_map['C'], label='C: Consultado'),
    Patch(facecolor=color_map['I'], label='I: Informado')
]

ax.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(-0.1, 1.1))

plt.show()
````

## 2.1 Cost of Poor Quality (CoPQ) ‚Äì alineado al Kano / VOC

El CoPQ se clasifica en cuatro categor√≠as:  
- **Costos de Fallas Internas (IFC):** p√©rdidas por ineficiencias dentro del proceso.  
- **Costos de Fallas Externas (EFC):** impacto cuando el cliente recibe retrasos o defectos.  
- **Costos de Evaluaci√≥n (AC):** costos para medir y controlar el proceso.  
- **Costos de Prevenci√≥n (PC):** inversi√≥n en acciones para evitar fallas.  

### Tabla de CoPQ basado en Kano/VOC

| # | VOC (cita corta)                                           | Categor√≠a Kano            | Tipo de CoPQ                 | Costo estimado / Impacto                             |
| - | ---------------------------------------------------------- | ------------------------- | ---------------------------- | --------------------------------------------------- |
| 1 | ‚ÄúTenemos sobrecostos por horas extra y contrataciones‚Äù (GE) | Must-be                   | **IFC** (sobrecostos directos) | ‚âà **S/ 61,440/a√±o** en horas extra.                 |
| 2 | ‚ÄúEl problema es la variaci√≥n en tiempos de ciclo‚Äù (GO)     | One-dimensional           | **IFC + EFC**                 | Variaciones generan cuellos de botella ‚Üí retrasos ‚Üí incumplimiento con clientes (riesgo contractual). |
| 3 | ‚ÄúLos equipos entran irregulares a test‚Äù (GST)              | One-dimensional           | **IFC**                       | Horas de espera + retrabajos ‚Üí baja productividad.   |
| 4 | ‚ÄúNo siempre hay personal suficiente en limpieza/pintado‚Äù (LT) | Must-be                   | **IFC + PC**                  | Costo de redistribuci√≥n de personal + horas no productivas. |
| 5 | ‚ÄúNecesitamos cumplir 4800 ONT en 15 d√≠as sin extras‚Äù (GE)  | Must-be / One-dimensional | **EFC**                       | Penalidad potencial por incumplimiento de contrato / p√©rdida de cliente (‚âà valor de contrato). |
| 6 | ‚ÄúNo tenemos visibilidad clara del ciclo‚Äù (PM s√≠ntesis)     | Attractive                | **AC + PC**                   | Inversi√≥n en dashboards / sistemas de monitoreo ‚Üí beneficio intangible alto (confianza, control). |

---

### Resumen Econ√≥mico del CoPQ
- **Costos de Fallas Internas (IFC):** ‚âà S/ 61,440/a√±o + p√©rdidas de productividad ocultas.  
- **Costos de Fallas Externas (EFC):** riesgo de incumplimiento contractual (potencialmente > S/ 200,000 si se afecta un contrato mayor).  
- **Costos de Evaluaci√≥n (AC):** inversi√≥n en medici√≥n (~ S/ 8,000 para dashboard).

## 2.4 Yield(Rendimiento)

````python
#  Calcular defectos y yield
# ----------------------------
# Defectos en cada etapa
df["Defectos_Limpieza"] = df["Cantidad_Limpieza"] - df["Cantidad_Test"]
df["Defectos_Test"] = df["Cantidad_Test"] - df["Cantidad_Encajado"]

# Rendimiento por etapa (Yield individual)
df["Yield_Limpieza"] = df["Cantidad_Test"] / df["Cantidad_Limpieza"]
df["Yield_Test"] = df["Cantidad_Encajado"] / df["Cantidad_Test"]
df["Yield_Encajado"] = df["Cantidad_Encajado"] / df["Cantidad_Encajado"]  # =1 siempre

# RTY (Rolled Throughput Yield)
df["RTY"] = df["Yield_Limpieza"] * df["Yield_Test"] * df["Yield_Encajado"]

# FPY (First Pass Yield): proporci√≥n de unidades que llegan buenas al final
FPY = df["Cantidad_Encajado"].sum() / df["Cantidad_Limpieza"].sum()

# RTY global: producto de los yields promedio
RTY_global = df["RTY"].mean()

print("üìä Resultados globales:")
print(f"First Pass Yield (FPY): {FPY:.4f} ({FPY*100:.2f}%)")
print(f"Rolled Throughput Yield (RTY): {RTY_global:.4f} ({RTY_global*100:.2f}%)")
````


##2.5 üìä An√°lisis de Calidad del Proceso

El script calcula m√©tricas clave de calidad para un proceso de manufactura, midiendo la eficiencia y el rendimiento bas√°ndose en la metodolog√≠a **Six Sigma**.

---

## üîç M√©tricas Calculadas

### 1. DPMO (Defectos por Mill√≥n de Oportunidades)
- El **DPMO** es la cantidad de defectos que ocurren por cada mill√≥n de oportunidades. En este caso, un defecto es una unidad que no se encaj√≥ despu√©s de la limpieza. Una oportunidad es cada unidad que pasa por el proceso de limpieza.
- Un DPMO bajo indica una alta calidad y un proceso eficiente.

### 2. Nivel Sigma
- El **Nivel Sigma** es una escala estandarizada que traduce el DPMO en un valor que representa la capacidad del proceso.
- Un valor de sigma m√°s alto significa que el proceso es m√°s consistente y produce menos defectos.
- El c√°lculo incluye un desplazamiento de **1.5 sigma**, un est√°ndar de la industria para reflejar la variabilidad a largo plazo de un proceso.

````python
from scipy.stats import norm

# ----------------------------
# 7. C√°lculo DPMO y Sigma
# ----------------------------
# Defectos totales (unidades perdidas en el proceso)
defectos_totales = (df["Cantidad_Limpieza"] - df["Cantidad_Encajado"]).sum()

# Oportunidades (cada unidad procesada en Limpieza)
oportunidades_totales = df["Cantidad_Limpieza"].sum()

# DPMO
DPMO = (defectos_totales / oportunidades_totales) * 1_000_000

# Nivel Sigma (aproximado, con desplazamiento de 1.5 sigma)
nivel_sigma = norm.ppf(1 - (DPMO / 1_000_000)) + 1.5

print("üìä M√©tricas de Calidad del Proceso")
print(f"Defectos Totales: {defectos_totales}")
print(f"Oportunidades Totales: {oportunidades_totales}")
print(f"DPMO: {DPMO:.2f}")
print(f"Nivel Sigma: {nivel_sigma:.2f}")
````

##2.6 Evaluaci√≥n de Capacidad del Proceso (Process Capability)

## üéØ Objetivo
Calcular los indicadores de capacidad del proceso (Cp, Cpk) para cada etapa clave del flujo (Recepci√≥n y Almacenado, Limpieza, Pintado, Test, Encajado). Esto permitir√° evaluar si los tiempos de ciclo cumplen de manera consistente con los l√≠mites de especificaci√≥n definidos.

## üìå Metodolog√≠a
1. **Definir l√≠mites de especificaci√≥n (USL/LSL)**:  
   - Cada proceso tiene un tiempo de ciclo esperado y una tolerancia (ejemplo: Pintado 8.5h ¬± 1h).  
   - Estos l√≠mites se usar√°n como referencia para calcular capacidad.

2. **Calcular estad√≠sticos descriptivos**:  
   - Media (\(\mu\)) y desviaci√≥n est√°ndar (\(\sigma\)) de cada proceso.  

3. **Calcular √≠ndices de capacidad**:  
   - **Cp = (USL - LSL) / (6œÉ)**  
     > Indica el potencial de capacidad del proceso.  
   - **Cpk = min((USL - Œº) / (3œÉ), (Œº - LSL) / (3œÉ))**  
     > Indica la capacidad real considerando el centrado respecto a los l√≠mites.  

4. **Interpretaci√≥n esperada**:  
   - **Cp < 1.0** ‚Üí El proceso no puede cumplir con las tolerancias.  
   - **Cp ‚âà 1.0** ‚Üí El proceso apenas cumple, requiere control estricto.  
   - **Cp ‚â• 1.33** ‚Üí El proceso es capaz y robusto.  
   - **Cpk ‚âà Cp** ‚Üí El proceso est√° bien centrado.  
   - **Cpk < Cp** ‚Üí El proceso est√° descentrado.  

## ‚úÖ Entregables
- Tabla con Cp y Cpk por proceso.  
- Identificaci√≥n de procesos fuera de control (Cp < 1 o Cpk muy bajo).  
- Recomendaciones para estabilizar y centrar el proceso.


````python
# @title
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Cargar dataset
df = pd.read_csv("https://raw.githubusercontent.com/AndreLir/MyDatasets/refs/heads/main/DataSet/DataLSSRefurbished.csv")

# Definir l√≠mites de especificaci√≥n (ejemplo)
spec_limits = {
    "TC_Horas_RyA": {"LSL": 26, "USL": 34},         # 50 ¬± 10
    "TC_Horas_Limpieza": {"LSL": 2.2, "USL": 3.6},  # rango esperado
    "TC_Horas_Test": {"LSL": 3, "USL": 4.5},      # rango esperado
    "TC_Horas_Encajado": {"LSL": 2.2, "USL": 3.4},  # rango esperado
}

def process_capability(series, LSL, USL, title):
    mu = series.mean()
    sigma = series.std(ddof=1)

    # Ensure sigma is not zero to avoid division by zero
    if sigma == 0:
        Cp = float('inf')
        Cpk = float('inf')
    else:
        Cp = (USL - LSL) / (6 * sigma)
        Cpk = min((USL - mu) / (3 * sigma), (mu - LSL) / (3 * sigma))

    # --- Histograma con curva normal ---
    plt.figure(figsize=(10,5))
    count, bins, ignored = plt.hist(series, bins=20, density=True, alpha=0.6, color="skyblue", edgecolor="black")
    plt.plot(bins, norm.pdf(bins, mu, sigma), 'r--', linewidth=2, label="Normal(Œº,œÉ)")
    plt.axvline(mu, color='blue', linestyle='--', linewidth=2, label=f"Media = {mu:.2f}")
    plt.title(f"Distribuci√≥n de {title}\nCp={Cp:.2f}, Cpk={Cpk:.2f}")
    plt.xlabel("Tiempo de ciclo (horas)")
    plt.ylabel("Densidad")
    plt.legend()
    plt.show()

    # --- Histograma con LSL y USL ---
    plt.figure(figsize=(10,5))
    plt.hist(series, bins=20, color="lightgreen", edgecolor="black", alpha=0.7)
    plt.axvline(LSL, color='red', linestyle='--', linewidth=2, label=f"LSL={LSL}")
    plt.axvline(USL, color='red', linestyle='--', linewidth=2, label=f"USL={USL}")
    plt.axvline(mu, color='blue', linestyle='--', linewidth=2, label=f"Media={mu:.2f}")
    plt.title(f"Capacidad del Proceso - {title}\nCp={Cp:.2f}, Cpk={Cpk:.2f}")
    plt.xlabel("Tiempo de ciclo (horas)")
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.show()

    return round(mu,2), round(sigma,2), round(Cp,3), round(Cpk,3)

# Ejecutar para cada proceso
results = {}
for col, limits in spec_limits.items():
    results[col] = process_capability(df[col], limits["LSL"], limits["USL"], col)

# Tabla resumen
capability_df = pd.DataFrame.from_dict(
    results,
    orient="index",
    columns=["Media", "Desv.Est", "Cp", "Cpk"]
)

display(capability_df)
````

##5.1 Evaluaci√≥n de Capacidad del Proceso (Cpk)

## üéØ Objetivo

El prop√≥sito de este an√°lisis es **comparar los nuevos valores de capacidad del proceso (Cpk)** con los obtenidos antes de la implementaci√≥n de mejoras.
De esta manera, podemos validar cuantitativamente **c√≥mo se comport√≥ el proyecto** en t√©rminos de reducci√≥n de variabilidad y alineamiento con los l√≠mites de especificaci√≥n.

---

## ‚öôÔ∏è Metodolog√≠a

1. Se carg√≥ el **dataset con 40 registros** representativos del proceso despu√©s de la mejora.
2. Se aplicaron los l√≠mites de especificaci√≥n definidos para cada etapa:

   * RyA: 6 ‚Äì 7 h
   * Limpieza: 2.21 ‚Äì 2.61 h
   * Test: 3 ‚Äì 3.4 h
   * Encajado: 2.2 ‚Äì 2.6 h
3. Se calcularon indicadores de capacidad:

   * **Cp**: mide el potencial del proceso respecto a los l√≠mites.
   * **Cpk**: mide qu√© tan centrado est√° el proceso dentro de los l√≠mites.

---

## üìà Resultados Esperados

* **Recepci√≥n y Almacenado (RyA):** fuerte reducci√≥n de tiempos (antes 27‚Äì36h ‚Üí ahora 6‚Äì7h).
* **Limpieza, Test y Encajado:** procesos estables y con variabilidad controlada.
* Visualizaci√≥n mediante histogramas y curvas normales para validar ajuste de datos.

---

## ‚úÖ Conclusi√≥n Parcial

El an√°lisis nos permitir√° confirmar que:

* Las mejoras implementadas redujeron tiempos de ciclo.
* Los procesos ahora operan dentro de l√≠mites controlados.
* El **Cpk aument√≥ significativamente**, demostrando que el proyecto cumpli√≥ su objetivo de estabilizar y hacer m√°s predecibles las operaciones.

---


````python
# @title
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Cargar dataset
df = pd.read_csv("https://raw.githubusercontent.com/AndreLir/MyDatasets/refs/heads/main/DataSet/DataLSSRefurbished_goal.csv")

# Definir l√≠mites de especificaci√≥n (ejemplo)
spec_limits = {
    "TC_Horas_RyA": {"LSL": 6, "USL": 7},         # 50 ¬± 10
    "TC_Horas_Limpieza": {"LSL": 2.21, "USL": 2.61},  # rango esperado
    "TC_Horas_Test": {"LSL": 3, "USL": 3.4},      # rango esperado
    "TC_Horas_Encajado": {"LSL": 2.2, "USL": 2.6
                          },  # rango esperado
}

def process_capability(series, LSL, USL, title):
    mu = series.mean()
    sigma = series.std(ddof=1)

    # Ensure sigma is not zero to avoid division by zero
    if sigma == 0:
        Cp = float('inf')
        Cpk = float('inf')
    else:
        Cp = (USL - LSL) / (6 * sigma)
        Cpk = min((USL - mu) / (3 * sigma), (mu - LSL) / (3 * sigma))

    # --- Histograma con curva normal ---
    plt.figure(figsize=(10,5))
    count, bins, ignored = plt.hist(series, bins=20, density=True, alpha=0.6, color="skyblue", edgecolor="black")
    plt.plot(bins, norm.pdf(bins, mu, sigma), 'r--', linewidth=2, label="Normal(Œº,œÉ)")
    plt.axvline(mu, color='blue', linestyle='--', linewidth=2, label=f"Media = {mu:.2f}")
    plt.title(f"Distribuci√≥n de {title}\nCp={Cp:.2f}, Cpk={Cpk:.2f}")
    plt.xlabel("Tiempo de ciclo (horas)")
    plt.ylabel("Densidad")
    plt.legend()
    plt.show()

    # --- Histograma con LSL y USL ---
    plt.figure(figsize=(10,5))
    plt.hist(series, bins=20, color="lightgreen", edgecolor="black", alpha=0.7)
    plt.axvline(LSL, color='red', linestyle='--', linewidth=2, label=f"LSL={LSL}")
    plt.axvline(USL, color='red', linestyle='--', linewidth=2, label=f"USL={USL}")
    plt.axvline(mu, color='blue', linestyle='--', linewidth=2, label=f"Media={mu:.2f}")
    plt.title(f"Capacidad del Proceso - {title}\nCp={Cp:.2f}, Cpk={Cpk:.2f}")
    plt.xlabel("Tiempo de ciclo (horas)")
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.show()

    return round(mu,2), round(sigma,2), round(Cp,3), round(Cpk,3)

# Ejecutar para cada proceso
results = {}
for col, limits in spec_limits.items():
    results[col] = process_capability(df[col], limits["LSL"], limits["USL"], col)

# Tabla resumen
capability_df = pd.DataFrame.from_dict(
    results,
    orient="index",
    columns=["Media", "Desv.Est", "Cp", "Cpk"]
)

display(capability_df)
````
